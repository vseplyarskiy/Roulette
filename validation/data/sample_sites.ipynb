```python
import pandas as pd
import dask.dataframe as dd
```

```python
from dask.distributed import Client
client = Client()
```

```python
df = dd.read_csv("synonymous_gnomAD_ukb_denovo/*part", sep = "\t")
```

```python
df.head()
```

### filter for sites with mutation rate estimates

```python
rate_column_name = ["mu", "Carlson_rate", "gnomAD_rate"]
for i in rate_column_name:
    df = df[df[i].isna() == False]
    df = df[df[i] != 0]
```

```python
df = df[df["mu_quality"] == "high"]
```

```python
mu_list = ["mu", "Carlson_rate", "gnomAD_rate"]
polymorphic_list = ["gnomADv2_polymorphic", "ukb_polymorphic", "denovo_genome", "denovo_exome"]
```

```python
df = df[polymorphic_list +mu_list]
```

```python
df = df.compute()
```

### output hash table

```python
df = df.fillna(0)
```

```python
df_sample = df.sample(n = len(df), replace = True)
```

```python
sample_size = 1000

for polymorphic in polymorphic_list:
    
#     print(polymorphic)
    
    for sample_num in range(sample_size):

        df_sample = df.sample(n = len(df), replace = True)

        for mu in mu_list:            
#             print(mu)
            
            df_grouped = pd.DataFrame(df_sample.groupby(mu).agg({polymorphic: ['size', 'sum']}))
            
            df_grouped = df_grouped.sort_values(mu)

            df_grouped.to_csv("sample_sites/" + polymorphic + "_" + mu + "_sample_" + str(sample_num) + "_binned.tsv", sep = "\t")
```

### output hash table for combined denovo

```python
df["denovo_combined"] = df["denovo_genome"] + df["denovo_exome"]
df["denovo_combined"] = df["denovo_combined"].replace(2, 1)
df["denovo_combined"] = df["denovo_combined"].astype(int)
```

```python
sample_size = 1000

for polymorphic in ["denovo_combined"]:
    
#     print(polymorphic)
    
    for sample_num in range(sample_size):

        df_sample = df.sample(n = len(df), replace = True)

        for mu in mu_list:            
#             print(mu)
            
            df_grouped = pd.DataFrame(df_sample.groupby(mu).agg({polymorphic: ['size', 'sum']}))
            
            df_grouped = df_grouped.sort_values(mu)

            df_grouped.to_csv("sample_sites/" + polymorphic + "_" + mu + "_sample_" + str(sample_num) + "_binned.tsv", sep = "\t")
```
